```python
from river import compose
from river import feature_extraction
from river import naive_bayes
```

```python
df = pd.read_json('combined_names_newest.json')


```

```python
df = df.head(10000)
```

```python
df.head(10)
```

```python
df['features'] = df['features'].astype('str')
```

```python
df.info()
```

```python
list = []

for i in df['features']:


    m = re.search(':(.+?),', i)
    if m:
        found = m.group(1)
        list.append(found)
```

```python
df['labels'] = df['labels'].astype('str')
```

```python
list1 = []

for i in df['labels']:


    m = re.search(':(.+?)]', i)
    if m:
        found = m.group(1)
        list1.append(found)
```

```python
X = pd.Series(list)
```

```python
Y = pd.Series(list1)
```

```python
model = compose.Pipeline(
     ("tokenize", feature_extraction.BagOfWords(lowercase=False)),
     ("nb", naive_bayes.BernoulliNB(alpha=1))
 )
```

```python
model = model.learn_many(X, Y)
```
